```python
import pandas as pd 

df = pd.read_csv("data-faskes-2019.csv")
df.info()
```

# Drop Column

```python
df = df.drop(['NoLink', 'Link', 'No'], axis=1)
df
```

# Drop Row `-`

```python
# print(f'Before drop row {len(df)}')
print(f'Before drop row : {len(df)}')

# Create a list of indexes to drop
listIndex = []
for idx, data in enumerate(df.values.tolist()):
    if '-' in data:
        listIndex.append(idx)

print(f'Invalid data total : {len(listIndex)}')

# Drop rows based on the list of indexes
df.drop(df.index[listIndex], inplace=True)

# Reset the index after dropping rows
df.reset_index(drop=True, inplace=True)

# Print the length of the DataFrame after dropping rows
print(f'After drop row : {len(df)}')
```

# Filter Tipe Faskes

```python
print(f'Before filter TipeFaskes {len(df)}')
df = df.query("TipeFaskes != 'Dokter Gigi'")
df = df.query("TipeFaskes != 'Apotek'")
df = df.query("TipeFaskes != 'Dokter Praktik Perorangan'")
print(f'After filter TipeFaskes {len(df)}')
```

# Unique Data Per-Column (Explore Untuk Normalisasi) & Cleansing

```python
df['TipeFaskes'].unique()
```

```python
df['Provinsi'].unique()
```

```python
df['KotaKab'] = df['KotaKab'].replace('Kode Faskes dan Alamat Rumah Sakit BPJS', '', regex = True)
df['KotaKab'] = df['KotaKab'].replace('di', '', regex = True)
df['KotaKab'] = df['KotaKab'].replace('\r\n', '', regex = True)
df['KotaKab'] = df['KotaKab'].replace('Kode Faskes dan Alamat Rumah Sakit            BPJS ', '', regex = True)
df['KotaKab'] = df['KotaKab'].replace('    ', ' ', regex = True)
df['KotaKab'] = df['KotaKab'].replace('   ', ' ', regex = True)
df['KotaKab'] = df['KotaKab'].replace('  ', ' ', regex = True)
df['KotaKab'] = df['KotaKab'].apply(lambda x: x.strip())
df['KotaKab'].unique()
```

```python
df['NamaFaskes'] = df['NamaFaskes'].replace('\r\n', '', regex=True)
df['NamaFaskes'] = df['NamaFaskes'].apply(lambda x: x.strip())
df['TelpFaskes'] = df['TelpFaskes'].replace('\r\n', '', regex=True)
df['TelpFaskes'] = df['TelpFaskes'].apply(lambda x: x.strip())
df['Provinsi'] = df['Provinsi'].replace('D I Yogyakarta', 'Daerah Istimewa Yogyakarta')
df['Provinsi'] = df['Provinsi'].replace('Dki Jakarta', 'DKI Jakarta')
df
```

# Add Column lat & lng

```python
detail = df['LatLongFaskes'].str.split('=|,', expand=True).iloc[:, 1:3]
detail = detail.rename(columns={1: 'Latitude', 2: 'Longitude'})
df['lat'] = detail['Latitude']
df['lng'] = detail['Longitude']
df
```

# Rename Column

```python
df = df.rename({
  "Provinsi": "province",
  "KotaKab": "city",
  "KodeFaskes": "kode_faskes",
  "NamaFakses": "facility",
  "LatLongFaskes": "gmap_url",
  "AlamatFaskes": "address",
  "TelpFaskes": "telephone",
  "TipeFaskes": 'type'
}, axis=1)
df
```

# To CSV

```python
dataProvince = df['province'].unique()
print(dataProvince)
pd.DataFrame({"province": dataProvince}).to_csv("result/provinces.csv", index=False)
print("Success save csv provinces")

dataCity = pd.DataFrame({
  "province": df["province"],
  "city": df["city"]
})


dataCity = dataCity.drop_duplicates(subset=['city'])
print(dataCity)
dataCity.to_csv("result/cities.csv", index=False)
print("Success save csv cities")


dataType = pd.DataFrame({
  "type": df['type'].unique()
})
print(dataType)
pd.DataFrame(dataType).to_csv("result/types.csv", index=False)
print("Success save csv types")

df.to_csv("result/facilities.csv", index=False)
print("Success save csv facilities")
```
